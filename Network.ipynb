```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
a=pd.read_csv('diabetes.csv')
a
```

```python
a=a[(a['Glucose']>20) & (a['BloodPressure']>20)& (a['SkinThickness']>10) & (a['Insulin']>40)& (a['BMI']>10)& (a['DiabetesPedigreeFunction']>0.1)]
a
```

```python
a.reset_index(inplace=True)
a
```

```python
data=a.iloc[:,2:10]
data
```

```python
#normalizasyon
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()

x_scaled = sc.fit_transform(data)
x_scaled
```

```python
labels=a['Outcome']
labels
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_scaled, labels, test_size=0.3, random_state=42)
```

```python
import random;
class Network:

    loss_values = [];

    def __init__(self) -> None:
        np.random.normal()
        self.w111 = np.random.normal()
        self.w112 = np.random.normal()
        self.w113 = np.random.normal()
        self.w114 = np.random.normal()
        self.w115 = np.random.normal()
        self.w116 = np.random.normal()
        self.w117 = np.random.normal()
        self.w118 = np.random.normal()
        self.b11 = np.random.normal()


        self.w121 = np.random.normal()
        self.w122 = np.random.normal()
        self.w123 = np.random.normal()
        self.w124 = np.random.normal()
        self.w125 = np.random.normal()
        self.w126 = np.random.normal()
        self.w127 = np.random.normal()
        self.w128 = np.random.normal()
        self.b12 = np.random.normal()

        self.w131 = np.random.normal()
        self.w132 = np.random.normal()
        self.w133 = np.random.normal()
        self.w134 = np.random.normal()
        self.w135 = np.random.normal()
        self.w136 = np.random.normal()
        self.w137 =np.random.normal()
        self.w138 = np.random.normal()
        self.b13 = np.random.normal()


        self.w211 = np.random.normal()
        self.w212 = np.random.normal()
        self.w213 = np.random.normal()
        self.b21 = np.random.normal()

        self.w311 = np.random.normal()
        self.b31 = np.random.normal()

        self.w321 = np.random.normal()
        self.b32 = np.random.normal()

        self.w411 = np.random.normal()
        self.w412 = np.random.normal()
        self.b41 = np.random.normal()


    def predict(self , data):


        h1 = self.sigmoid((self.w111 * data[0]) + (self.w112 * data[1])
                            +(self.w113 * data[2]) + (self.w114 * data[3])
                            +(self.w115 * data[4]) + (self.w116 * data[5])
                            +(self.w117 * data[6])+ (self.w118 * data[7])
                            +(self.b11))



        # h2 nöronunun değeri
        h2 = self.sigmoid((self.w121 * data[0]) + (self.w122 * data[1])
                            +(self.w123 * data[2]) + (self.w124 * data[3])
                            +(self.w125 * data[4]) + (self.w126 * data[5])
                            +(self.w127 * data[6])+ (self.w128 * data[7])
                            +(self.b12)
                        )




        h3 = self.sigmoid((self.w131 * data[0]) + (self.w132 * data[1])
                            +(self.w133 * data[2]) + (self.w134 * data[3])
                            +(self.w135 * data[4]) + (self.w136 * data[5])
                            +(self.w137 * data[6])+ (self.w138 * data[7])
                            +(self.b13)
                        )




        # Tahmin değeri 01 nöronun değeri
        h4 = self.sigmoid((self.w211 * h1 ) + (self.w212 * h2 ) + (self.w213 * h3) + (self.b21))





        h5 = self.sigmoid((self.w311 * h4) + (self.b31))

        h6 = self.sigmoid((self.w321 * h4) + (self.b32))

        o1 = self.sigmoid((self.w411 * h5) + (self.w412 * h6) + (self.b41));


        return o1

    def train(self , data , labels, epochs, learning_rate):

        for epoch in range(epochs):

            for x, y in zip(data , labels):

                h1 = self.sigmoid((self.w111 * x[0]) + (self.w112 * x[1])
                            +(self.w113 * x[2]) + (self.w114 * x[3])
                            +(self.w115 * x[4]) + (self.w116 * x[5])
                            +(self.w117 * x[6])+ (self.w118 * x[7])
                            +(self.b11)
                         )



                # h2 nöronunun değeri
                h2 = self.sigmoid((self.w121 * x[0]) + (self.w122 * x[1])
                                    +(self.w123 * x[2]) + (self.w124 * x[3])
                                    +(self.w125 * x[4]) + (self.w126 * x[5])
                                    +(self.w127 * x[6])+ (self.w128 * x[7])
                                    +(self.b12)
                                )



                h3 = self.sigmoid((self.w131 * x[0]) + (self.w132 * x[1])
                                    +(self.w133 * x[2]) + (self.w134 * x[3])
                                    +(self.w135 * x[4]) + (self.w136 * x[5])
                                    +(self.w137 * x[6])+ (self.w138 * x[7])
                                    +(self.b13)
                                )


                # Tahmin değeri 01 nöronun değeri
                h4 = self.sigmoid((self.w211 * h1 ) + (self.w212 * h2 ) + (self.w213 * h3) + (self.b21))



                h5 = self.sigmoid((self.w311 * h4) + (self.b31))



                h6 = self.sigmoid((self.w321 * h4) + (self.b32))

                o1 = self.sigmoid((self.w411 * h5) + (self.w412 * h6) + (self.b41));





                error = y - o1;


                o1_error_measure = self.sigmoid_der(o1) * error;



                h6_error_measure = self.sigmoid_der(h6) * (o1_error_measure * self.w412);



                h5_error_measure = self.sigmoid_der(h5) * (o1_error_measure * self.w411);



                h4_error_measure = self.sigmoid_der(h4) * ((h6_error_measure * self.w321) + (h5_error_measure * self.w311))



                h3_error_measure = self.sigmoid_der(h3) * (h4_error_measure * self.w213);



                h2_error_measure = self.sigmoid_der(h2) * (h4_error_measure * self.w212);



                h1_error_measure = self.sigmoid_der(h1) * (h4_error_measure * self.w211);







                # Wxyz += dWxyz


                self.w411 += learning_rate * o1_error_measure * h5;
                self.w412 += learning_rate * o1_error_measure * h6;


                self.w311 += learning_rate * h5_error_measure * h4;
                self.w321 += learning_rate * h6_error_measure * h4;


                self.w211 += learning_rate * h4_error_measure * h1;
                self.w212 += learning_rate * h4_error_measure * h2;
                self.w213 += learning_rate * h4_error_measure * h3;



                self.w111 += learning_rate * h1_error_measure * x[0];
                self.w112 += learning_rate * h1_error_measure * x[1];
                self.w113 += learning_rate * h1_error_measure * x[2];
                self.w114 += learning_rate * h1_error_measure * x[3];
                self.w115 += learning_rate * h1_error_measure * x[4];
                self.w116 += learning_rate * h1_error_measure * x[5];
                self.w117 += learning_rate * h1_error_measure * x[6];
                self.w118 += learning_rate * h1_error_measure * x[7];
                #-
                self.w121 += learning_rate * h2_error_measure * x[0];
                self.w122 += learning_rate * h2_error_measure * x[1];
                self.w123 += learning_rate * h2_error_measure * x[2];
                self.w124 += learning_rate * h2_error_measure * x[3];
                self.w125 += learning_rate * h2_error_measure * x[4];
                self.w126 += learning_rate * h2_error_measure * x[5];
                self.w127 += learning_rate * h2_error_measure * x[6];
                self.w128 += learning_rate * h2_error_measure * x[7];
                #-
                self.w131 += learning_rate * h3_error_measure * x[0];
                self.w132 += learning_rate * h3_error_measure * x[1];
                self.w133 += learning_rate * h3_error_measure * x[2];
                self.w134 += learning_rate * h3_error_measure * x[3];
                self.w135 += learning_rate * h3_error_measure * x[4];
                self.w136 += learning_rate * h3_error_measure * x[5];
                self.w137 += learning_rate * h3_error_measure * x[6];
                self.w138 += learning_rate * h3_error_measure * x[7];


                self.b41 += learning_rate * o1_error_measure;
                self.b31 += learning_rate * h5_error_measure;
                self.b32 += learning_rate * h6_error_measure;
                self.b21 += learning_rate * h4_error_measure;
                self.b11 += learning_rate * h1_error_measure;
                self.b12 += learning_rate * h2_error_measure;
                self.b13 += learning_rate * h3_error_measure;

            if(epoch % 300 == 0):
              predictions = np.apply_along_axis(self.predict ,1, data)
              loss = self.mse_loss(labels , predictions)
              print("Devir %d loss: %.7f" % (epoch, loss))
              self.loss_values.append(loss);


        predictions = np.apply_along_axis(self.predict ,1, data)
        loss = self.mse_loss(labels , predictions)
        print("Devir %d loss: %.7f" % (epoch, loss))


    def sigmoid(self , x):
        return (1/(1+np.exp(-x)))

    def sigmoid_der(self , x):
        return x * (1 - x)

    def mse_loss(self , y_real , y_prediction):
        return ((y_real - y_prediction) ** 2).mean()
```

```python
myNetwork = Network();
```

```python
myNetwork.train(X_train, y_train, 10000, 0.05);
```

```python
for x, y in zip(X_test , y_test):
    value = myNetwork.predict(x);
    print("Predict -> {} Gerçek Değer -> {}".format(value, y));
```

```python
all_prediction =  [];
for x in X_test:
    if(myNetwork.predict(x) > 0.5):
        all_prediction.append(1)
    else:
        all_prediction.append(0);
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, all_prediction)
cm.T
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, all_prediction));
```
